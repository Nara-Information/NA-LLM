# OpenAPI Playground

A playground notebook to inspect the OepnAPI from `data.go.kr`

```python
import json

from pprint import pprint

import yaml
import requests
```

```python
with open('creds.yaml') as f:
    creds = yaml.load(f, yaml.Loader)
keys = creds['OpenAPI']
keys
```

```python
import requests

URL = "http://apis.data.go.kr/1140100/CivilPolicyQnaService/PolicyQnaList"

params = {
    'serviceKey': keys['key-decoded'],
    'firstIndex': '1',
    'recordCountPerPage': '100',
    'type': '1',
    'regFrom': '20230101',
    'regTo': '20230531'
}

response = requests.get(URL, params=params)
try:
    content = json.loads(response.content)
except:
    content = response.content
pprint(content)
```

```python
faqNos = [d['faqNo'] for d in content['resultList']]
faqNos
```
